```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,125,5)
x
```

```python
y1 = np.random.uniform(-10,2,len(x))
y2 = np.random.uniform(0,20,len(x))
```

```python
fig,ax1 = plt.subplots()
```

```python
fig,ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')
```

```python
fig,ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')


ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2')

plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

ax1.plot(x, y1)

ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2')

ax2.plot(x, y2)

plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

ax1.plot(x, y1,color = 'royalblue')

ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2')

ax2.plot(x, y2,color = 'tomato')

plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

ax1.plot(x, y1,color = 'royalblue',marker = 'o',markersize = 8, linestyle = '-')

ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2')

ax2.plot(x, y2,color = 'tomato',marker = 's',markersize = 8, linestyle = '-.')

plt.show()
```

```python
fig,ax1 = plt.subplots()

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

ax1.plot(x, y1,color = 'royalblue',marker = 'o',markersize = 8, linestyle = '-')

ax1.set_ylim([-10,10])


ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2')

ax2.plot(x, y2,color = 'tomato',marker = 's',markersize = 8, linestyle = '-.')

ax2.set_ylim([-10,20])

plt.show()
```

```python
fig,ax1 = plt.subplots(figsize = (8,5))

ax1.set_xlabel('X')
ax1.set_ylabel('Y1')

ax1.plot(x, y1,color = 'royalblue',marker = 'o',markersize = 8, linestyle = '-')

ax1.set_ylim([-10,10])


ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2')

ax2.plot(x, y2,color = 'tomato',marker = 's',markersize = 8, linestyle = '-.')

ax2.set_ylim([-10,20])

plt.show()
```

```python
plt.rcParams.update({'font.size':14})
```

```python
fig,ax1 = plt.subplots(figsize = (12,6))

ax1.set_xlabel('X')

color1 = 'royalblue'
ax1.set_ylabel('Y1',color = color1)
ax1.plot(x, y1,color = color1,marker = 'o',markersize = 8, linestyle = '-')

ax1.set_ylim([-10,10])
ax1.tick_params(axis = 'y',labelcolor = color1)

color2 = 'tomato'
ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2',color = color2)

ax2.plot(x, y2,color = color2,marker = 's',markersize = 8, linestyle = '-.')

ax2.tick_params(axis = 'y',labelcolor = color2)
ax2.set_ylim([-10,20])

plt.show()
```

```python
fig,ax1 = plt.subplots(figsize = (12,6))

ax1.set_xlabel('X')
ax1.set_ylabel('Y1',color = color1)
color1 = 'royalblue'
ax1.plot(x, y1,color = color1,marker = 'o',markersize = 8, linestyle = '-')

ax1.set_ylim([-10,10])
ax1.tick_params(axis = 'y',labelcolor = color1)

color2 = 'tomato'
ax2 = ax1.twinx() # compartilhar o mesmo eixo x
ax2.set_ylabel('Y2',color = color2)

ax2.plot(x, y2,color = color2,marker = 's',markersize = 8, linestyle = '-.')

ax2.tick_params(axis = 'y',labelcolor = color2)
ax2.set_ylim([-10,20])

plt.savefig('Grafico.png',format='png',dpi = 600)
plt.show()
```
